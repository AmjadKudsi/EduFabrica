<br>
The Iris dataset was used in R.A. Fisher's classic 1936 paper, The Use of Multiple Measurements in Taxonomic Problems, and can also be found on the UCI Machine Learning Repository.<br><br>

It includes three iris species with 50 samples each as well as some properties about each flower. One flower species is linearly separable from the other two, but the other two are not linearly separable from each other.<br> <br>

The columns in this dataset are:<br><br>

- Id
- SepalLengthCm
- SepalWidthCm
- PetalLengthCm
- PetalWidthCm
- Species

```python
# required packages
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = 'Iris.csv'
data = pd.read_csv(dataset,index_col=['Id'])
data.head(10)
```

```python
plt.figure(figsize=(12,5))
plt.title("Comparisons")
df=data.drop(columns=['Species'])
sns.lineplot(data=df)
plt.ylabel("length")
```

```python
plt.figure(figsize=(12,5))
plt.title("Sepal length data")
sns.barplot(x=data.Species,y=data.SepalLengthCm)
```

```python
plt.figure(figsize=(12,7))
plt.title("Sepal length vs Petal length")
sns.scatterplot(x=data.SepalLengthCm,y=data.PetalLengthCm)
```

```python
plt.figure(figsize=(12,8))
plt.title("Petal lenth vs Sepal length")
sns.scatterplot(x=data.SepalLengthCm,y=data.PetalLengthCm,hue=data.Species)
```

```python
plt.figure(figsize=(12,6))
plt.title("Sepal length vs sepal width")
sns.regplot(x=data.SepalLengthCm,y=data.SepalWidthCm)
```

```python
sns.lmplot(x='PetalLengthCm',y='PetalWidthCm',hue='Species',data=data)
```

```python
sns.lmplot(x='SepalLengthCm',y='PetalLengthCm',hue='Species',data=data)
```

```python
sns.swarmplot(x=data.SepalLengthCm,y=data.Species)
```

```python
plt.figure(figsize=(12,7))
plt.title("Variance of Sepal Length")
sns.kdeplot(data=data.SepalLengthCm,shade=True)
plt.xlabel("length")
plt.ylabel('Variance')
```

```python
plt.figure(figsize=(12,8))
plt.title("Variance of Sepal Length")
sns.distplot(a=data.SepalWidthCm,kde=False)
plt.ylabel("limit")
```

```python
sns.jointplot(x=data.SepalLengthCm,y=data.PetalLengthCm,kind='kde')
print("Variance of PetalLength is very high it may be result in overfiting of model")
```
